```python
#Ok, after you've located yourself in the same folder, we can start analyzing the data. 
#You need to have your excel files in the same folder. 
#To read the excel file: 
import pandas as pd
#The excel file has different sheets, but right now we will only use the one with quantification data.
df = pd.read_excel("Name_of_your_file.xlsx", sheet_name="Quantification") #Sheet name can change if you're not using the same software (MAPDP)
```

```python
#We create the groups of each treatment
Treatment_1 = ['Area Sample 10', 'Area Sample 11', 'Area Sample 12']
Treatment_2 = ['Area Sample 9', 'Area Sample 8', 'Area Sample 7']
Treatment_3 = ['Area Sample 6', 'Area Sample 5', 'Area Sample 4']
Treatment_4 = ['Area Sample 3', 'Area Sample 2', 'Area Sample 1']
```

```python
#We now create a new dataframe for each treatment we are comparing here, or the dataframe itself if you're not comparing different treatments. 
#This is an example
df_Treatment_1 = df[['Peptide sequence']+ Treatment_1].copy()
df_Treatment_1.head()
```

```python
#We continue with the rest
df_Treatment_2 = df[['Peptide sequence']+ Treatment_2].copy()
df_Treatment_3 = df[['Peptide sequence']+ Treatment_3].copy()
df_Treatment_4 = df[['Peptide sequence']+ Treatment_4].copy()
```

```python
#We need to know, for each peptide, in how many of the replicates it appears. 
#This helps us to filter between peptides with actual signals (if it appears in at least 2/3 replicates), 
#and peptides that might appear by error.
df_Treatment_1["Count_if"] = (df_Treatment_1[Treatment_1] > 0).sum(axis=1)
df_Treatment_2["Count_if"] = (df_Treatment_2[Treatment_2] > 0).sum(axis=1)
df_Treatment_3["Count_if"] = (df_Treatment_3[Treatment_3] > 0).sum(axis=1)
df_Treatment_4["Count_if"] = (df_Treatment_4[Treatment_4] > 0).sum(axis=1)

#df_Treatment_1.head()
```

```python
#Then, we need to keep only the peptides that appear at least 2 times in the dataset. 
#For that, we use the following:
filt_df_Treatment_1 = df_Treatment_1[df_Treatment_1["Count_if"] >= 2].copy()
filt_df_Treatment_2 = df_Treatment_2[df_Treatment_2["Count_if"] >= 2].copy()
filt_df_Treatment_3 = df_Treatment_3[df_Treatment_3["Count_if"] >= 2].copy()
filt_df_Treatment_4 = df_Treatment_4[df_Treatment_4["Count_if"] >= 2].copy()
```

```python
#You can check if the filtering was ok using this line:
filt_df_Treatment_1.head()
```

```python
#Then, we need to replace the 0's that appear in our dataset with the lower value in each of our treatments. This will help us have a more exact CV%
#and to better understand the replicability of our data. Not doing it risks higher CV%s and worst understanding of the results. 
#For this step, we first need to define the lower signal value per treatment.
min_signal_Treatment_1 = filt_df_Treatment_1[Treatment_1][filt_df_Treatment_1[Treatment_1] > 0].min().min()
#print(min_signal_Treatment_1)
min_signal_Treatment_2 = filt_df_Treatment_2[Treatment_2][filt_df_Treatment_2[Treatment_2] > 0].min().min()
min_signal_Treatment_3 = filt_df_Treatment_3[Treatment_3][filt_df_Treatment_3[Treatment_3] > 0].min().min()
min_signal_Treatment_4 = filt_df_Treatment_4[Treatment_4][filt_df_Treatment_4[Treatment_4] > 0].min().min()
```

```python
#Then, we will do the replacement of the 0's.
filt_df_Treatment_1[Treatment_1] = filt_df_Treatment_1[Treatment_1].replace(0, min_signal_Treatment_1)
#filt_df_Treatment_1.head()
filt_df_Treatment_2[Treatment_2] = filt_df_Treatment_2[Treatment_2].replace(0, min_signal_Treatment_2)
filt_df_Treatment_3[Treatment_3] = filt_df_Treatment_3[Treatment_3].replace(0, min_signal_Treatment_3)
filt_df_Treatment_4[Treatment_4] = filt_df_Treatment_4[Treatment_4].replace(0, min_signal_Treatment_4)

#Missing or zero intensity values are imputed using the lowest detected intensity within each treatment.
```

```python
#Here, we do the desired calculations. Mean of intensity, Standard deviation, and CV%.
#This is per peptide found in each dataframe. 
filt_df_Treatment_1["Mean"] = filt_df_Treatment_1[Treatment_1].mean(axis=1)
filt_df_Treatment_1["SD"] = filt_df_Treatment_1[Treatment_1].std(axis=1)
filt_df_Treatment_1["CV"] = (filt_df_Treatment_1["SD"] / filt_df_Treatment_1["Mean"]) * 100
filt_df_Treatment_1[["Peptide sequence", "Mean", "SD", "CV"]].head()
```

```python
#For treatment 2
filt_df_Treatment_2["Mean"] = filt_df_Treatment_2[Treatment_2].mean(axis=1)
filt_df_Treatment_2["SD"] = filt_df_Treatment_2[Treatment_2].std(axis=1)
filt_df_Treatment_2["CV"] = (filt_df_Treatment_2["SD"] / filt_df_Treatment_2["Mean"]) * 100
filt_df_Treatment_2[["Peptide sequence", "Mean", "SD", "CV"]].head()

#For treatment 3
filt_df_Treatment_3["Mean"] = filt_df_Treatment_3[Treatment_3].mean(axis=1)
filt_df_Treatment_3["SD"] = filt_df_Treatment_3[Treatment_3].std(axis=1)
filt_df_Treatment_3["CV"] = (filt_df_Treatment_3["SD"] / filt_df_Treatment_3["Mean"]) * 100
filt_df_Treatment_3[["Peptide sequence", "Mean", "SD", "CV"]].head()

#For treatment 4
filt_df_Treatment_4["Mean"] = filt_df_Treatment_4[Treatment_4].mean(axis=1)
filt_df_Treatment_4["SD"] = filt_df_Treatment_4[Treatment_4].std(axis=1)
filt_df_Treatment_4["CV"] = (filt_df_Treatment_4["SD"] / filt_df_Treatment_4["Mean"]) * 100
filt_df_Treatment_4[["Peptide sequence", "Mean", "SD", "CV"]].head()
```

```python
#Finally, we create the summary table with our final results. This gives us per treatment:
#Ammount of MAPs identified. 
#Average CV%
#Median CV%

summary_table = pd.DataFrame({
    "Treatment": [
        "Treatment 1",
        "Treatment 2",
        "Treatment 3",
        "Treatment 4"
    ],
    "MAPs #": [
        len(filt_df_Treatment_1),
        len(filt_df_Treatment_2),
        len(filt_df_Treatment_3),
        len(filt_df_Treatment_4)
    ],
    "Average CV%": [
        filt_df_Treatment_1["CV"].mean(),
        filt_df_Treatment_2["CV"].mean(),
        filt_df_Treatment_3["CV"].mean(),
        filt_df_Treatment_4["CV"].mean()
    ],
    "Median CV%": [
        filt_df_Treatment_1["CV"].median(),
        filt_df_Treatment_2["CV"].median(),
        filt_df_Treatment_3["CV"].median(),
        filt_df_Treatment_4["CV"].median()
    ]
})
```

```python
#If you want to round decimals to have only 2. 
summary_table["Average CV%"] = summary_table["Average CV%"].round(2)
summary_table["Median CV%"] = summary_table["Median CV%"].round(2)
```
